```python
# Convertir booking_status a valores numéricos (1 = Canceled, 0 = Not_Canceled)
df["booking_status_numeric"] = df["booking_status"].map({"Canceled": 1, "Not_Canceled": 0})

# Calcular la correlación entre las variables numéricas y booking_status_numeric
correlation_matrix = df.corr(numeric_only=True)["booking_status_numeric"].sort_values(ascending=False)

# Mostrar las 10 variables más correlacionadas con la cancelación
print(correlation_matrix.head(10))
```

```python

# Configuración general del estilo
sns.set(style="whitegrid")

#Gráfico Univariado: Histograma de Lead Time
plt.figure(figsize=(8, 5))
sns.histplot(df["lead_time"], bins=30, kde=True, color="blue")
plt.title("Distribución del Lead Time")
plt.xlabel("Lead Time (días)")
plt.ylabel("Frecuencia")
plt.show()

```

 El gráfico muestra como se distribuyen los tiempos de anticipación de una reserva. Se puede observar una distribución sesgada hacia la derecha. La mayoria de las reservas se realízan con un corto tiempo de anticipación, a medida que el lead time aumenta la cantidad de reservas disminuye. Existen algunos valores atípicos con mas de 300 días de anticipación.
 Existe una gran cantidad de reservas con menos de 15 dias de anticipación, lo que significa que muchas personas hacen reservas a ultima hora.

```python
#Gráfico Bivariado: Boxplot de Lead Time según Booking Status
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["booking_status"], y=df["lead_time"], hue=df["booking_status"], palette="coolwarm", legend=False)
plt.title("Lead Time vs Booking Status")
plt.xlabel("Estado de Reserva")
plt.ylabel("Lead Time (días)")
plt.show()
```

```python
# --- 3. Gráfico Multivariado: Scatterplot de Lead Time vs Precio de Habitación, coloreado por Booking Status ---
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="lead_time", y="avg_price_per_room", hue="booking_status", alpha=0.6, palette="viridis")
plt.title("Lead Time vs Precio por Habitación (Segmentado por Cancelación)")
plt.xlabel("Lead Time (días)")
plt.ylabel("Precio Promedio por Habitación")
plt.legend(title="Booking Status")
plt.show()
```
